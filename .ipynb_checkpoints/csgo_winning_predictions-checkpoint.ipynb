```python
def save_obj(obj, name):
    os.makedirs("obj/", exist_ok=True)
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
csgo_data = pd.concat([csgo_data, pd.get_dummies(csgo_data['map'], prefix='map')],axis=1)

csgo_data.drop(['map', 'map_de_cache'], axis=1, inplace=True)
csgo_data
```

```python
csgo_data['round_winner'] = rw_col.astype(int)
csgo_data['round_winner']
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(csgo_data, csgo_data['round_winner']):
    strat_train_set = csgo_data.loc[train_index]
    strat_test_set = csgo_data.loc[test_index]
```

```python
X_train_raw = strat_train_set.drop('round_winner', axis=1)
y_train = strat_train_set['round_winner'].copy()
```

```python
from sklearn.preprocessing import MinMaxScaler

X_train = MinMaxScaler().fit_transform(X_train_raw)
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict

forest_clf = RandomForestClassifier(random_state = 42)
y_probas_forest = cross_val_predict(forest_clf, X_train, y_train, cv = 3, method = "predict_proba")
y_probas_forest
```

```python
from sklearn.metrics import precision_recall_curve

def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    fig = plt.figure(figsize=(10, 7))
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.xlabel("Threshold")
    plt.legend(loc = 'best')
    plt.grid()

y_scores_forest = y_probas_forest[:, 1]
precisions_forest, recalls_forest, thresholds_forest = precision_recall_curve(y_train, y_scores_forest)
plot_precision_recall_vs_threshold(precisions_forest, recalls_forest, thresholds_forest)
plt.show()
```

```python
def plot_prec_vs_recall(precision, recall, label = None):
    fig = plt.figure(figsize=(10, 7))
    plt.plot(recall, precision, linewidth = 2, label = label)
    plt.ylabel("Precision")
    plt.xlabel("Recall")
    plt.legend(loc = 'best')
    plt.grid()
    
plot_prec_vs_recall(precisions_forest, recalls_forest, label = "forest")
plt.legend(loc="lower left")
plt.show()
```

```python
best_prod = 0
i_max = -1
for i in range(len(precisions_forest)):
    prec = precisions_forest[i]
    rec = recalls_forest[i]
    if (best_prod < 2 * (prec * rec) / (prec + rec)):
        best_prod = 2 * (prec * rec) / (prec + rec)
        i_max = i

best_thr = thresholds_forest[i_max]
best_thr
```

```python
def predict_by_proba(X_, model, threshold):
    y_probas = model.predict_proba(X_)
    y_scores = y_probas[:, 1]
    y_pred = (y_scores >= threshold)
    return y_pred

def predict_by_decision_func(X_, model, threshold):
    y_scores = model.decision_function(X_)
    y_pred = (y_scores >= threshold)
    return y_pred
```
